```python
import numpy as np
import pandas as pd

!pip install matplotlib
import matplotlib.pyplot as plt
import matplotlib as mpl

spotify = pd.read_csv("spotify-2023.csv", encoding="ISO-8859-1")
spotify.head()
```

```python
spotify.columns = ["musicaNome", "artista", "qArtista", "anoMusica", "mesMusica", "mesDia", "playlistSpot", "playlistSpotRanking", "streams", "playlistApple", "playlistAppleRanking", "playlistDeezer", "playlistDeezerRanking", "playlistShazamRanking", "bpm", "notaMusica", "mode", "dancante", "valence", "energia", "acustico", "instrumental", "vivacidade", "speechiness"]
spotify.head()
```

```python
listaEd = spotify.loc[[138, 179, 215, 255, 274, 429, 475, 595, 652, 685, 699, 751, 761, 925]]
listaEd
```

```python
mediaEd = listaEd['streams'].astype(float).mean()
mediaEd
```

```python
mediaTaylor = spotify[spotify['artista'].str.contains('Taylor Swift')]['streams'].astype(float).mean()
mediaTaylor
```

```python
art = ['Ed Sheeran', 'Taylor Swift']
media = [mediaEd, mediaTaylor]

colors = {
      'Ed Sheeran': '#41E9FF',
      'Taylor Swift': '#5C1914'
}

plt.bar(art, media, color=[colors[art] for art in art])
plt.ylabel('Média dos streams')
plt.title('Comparação da média de Streams estre Taylor Swift e Ed Sheeran')
plt.show()
```

```python
listaEd = listaEd.sort_values(by='streams', ascending=False)

listaEd.loc[:, 'streams'] = pd.to_numeric(listaEd['streams'], errors='coerce')

mus = ['Shape of You', 'Perfect', 'Thinking Out Loud', 'Photograph', 'Bad Habits']

colors = {
      'Shape of You': '#FFFB02',
      'Perfect': '#FF2500',
      'Thinking Out Loud': '#41E9FF',
      'Photograph': '#36FF65',
      'Bad Habits': '#FF9A30'
}

top_n = 5
listaEdS = listaEd.head(top_n)

plt.barh(listaEdS['musicaNome'], listaEdS['streams'], color=[colors[mus] for mus in mus])
plt.xlabel('Total de Streams')
plt.ylabel('Música')
plt.title('Músicas mais populares do Ed Sheeran por Streams')
plt.gca().invert_yaxis()
plt.show()
```

```python
listaEd = listaEd.sort_values(by='dancante', ascending=False)

listaEd.loc[:, 'dancante'] = pd.to_numeric(listaEd['dancante'], errors='coerce')

mus = ['Peru', 'Shape of You', 'Bad Habits', 'Thinking Out Loud', 'Eyes Closed']

colors = {
       'Peru': '#FFFB02',
       'Shape of You': '#FF2500',
       'Bad Habits': '#41E9FF',
       'Thinking Out Loud': '#36FF65',
       'Eyes Closed': '#FF9A30'
}

top_n = 5
listaEdS = listaEd.head(top_n)

plt.barh(listaEdS['musicaNome'], listaEdS['dancante'], color=[colors[mus] for mus in mus])
plt.xlabel('Valores')
plt.ylabel('Músicas')
plt.title('Top 5 músicas mais dançantes do Ed Sheeran')
plt.gca().invert_yaxis()
plt.show()
```

```python
listaTw = spotify[spotify['artista'].str.contains('The Weeknd')]['energia'].astype(float).mean()
listaTaylor = spotify[spotify['artista'].str.contains('Taylor Swift')]['energia'].astype(float).mean()
listaHarry = spotify[spotify['artista'].str.contains('Harry Styles')]['energia'].astype(float).mean()
listaEdSheeran = spotify[spotify['artista'].str.contains('Ed Sheeran')]['energia'].astype(float).mean()

ener = ['Ed Sheeran', 'Taylor Swift', 'Harry Styles', 'The Weeknd']
media = [listaEdSheeran, listaTaylor, listaHarry, listaTw]


colors = {
       'Ed Sheeran': '#41E9FF',
       'Taylor Swift': '#5C1914',
       'Harry Styles': '#9a7ee0',
       'The Weeknd': '#e649a2'
}


plt.barh(ener, media, color=[colors[ener] for ener in ener])
plt.xlabel('Média de energia das músicas')
plt.ylabel('Artista')
plt.title('Comparação de energia dentre artistas')
plt.gca().invert_yaxis()
plt.show()
```

```python
mediaAm = spotify[spotify['artista'].str.contains('Arctic Monkeys')]['acustico'].astype(float).mean()
mediaEdSheeran = spotify[spotify['artista'].str.contains('Ed Sheeran')]['acustico'].astype(float).mean()

art = ['Arctic Monkeys', 'Ed Sheeran']
media = [mediaAm, mediaEdSheeran]

colors = {
      'Ed Sheeran': '#41E9FF',
      'Arctic Monkeys': 'black'
}

plt.bar(art, media, color=[colors[art] for art in art])
plt.ylabel('Valores')
plt.title('Qual estilo é mais acústico (rock VS pop)')
plt.show()
```
